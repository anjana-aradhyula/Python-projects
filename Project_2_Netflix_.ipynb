#**2. Year in which highest number of TV shows and movies were released, along with bar graph**

```python
df.dtypes ##datatypes of each column
```

```python
## convert release date into DateTime format
df['Date_N'] = pd.to_datetime(df['Release_Date'])
```

```python
df.dtypes ##to check converted datatype
```

```python
##to count the number of times each year occurs in the Date_N column
df['Date_N'].dt.year.value_counts()
```

```python
##plot bar graph for number od times each year occurs in the Date_N column
df['Date_N'].dt.year.value_counts().plot(kind='bar')
```

#**3. Number of movies and TV shows, and a bar graph to show it**

#**All the movies released in the year 2020**

```python
##creating a new column just for release years
df['Year']=df['Date_N'].dt.year
```

```python
##to get movies released in 2020
df[ (df['Category']=='Movie') & (df['Year']==2020) ]
```

#**5. Show only the titles of TV shows released in India only**

#**6. Top 10 directors who gave the highest number of TV shows and movies to Netflix**
